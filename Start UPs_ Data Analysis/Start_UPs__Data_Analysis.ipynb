# 1. Data Types
The dataset contains:

* Numerical (Continuous):

  * R&D Spend, Administration, Marketing Spend, Profit

* Categorical (Nominal):

  * State (New York, California, Florida)


# 2. Five Visualization Types
* Scatter Plot

* Bar Chart

* Box Plot

* Histogram

* Heatmap (Correlation Matrix)

```python
# Set style
sns.set_style("whitegrid")
plt.figure(figsize=(15, 10))

# 1. Scatter Plot: R&D Spend vs. Profit
plt.subplot(2, 3, 1)
sns.scatterplot(x='R&D Spend', y='Profit', data=df, hue='State', palette='viridis')
plt.title("R&D Spend vs. Profit")

# 2. Bar Chart: Average Profit by State
plt.subplot(2, 3, 2)
sns.barplot(x='State', y='Profit', data=df, ci=None, palette='Set2')
plt.title("Average Profit by State")

# 3. Box Plot: Profit Distribution
plt.subplot(2, 3, 3)
sns.boxplot(y='Profit', data=df, color='skyblue')
plt.title("Profit Distribution")

# 4. Histogram: Marketing Spend Distribution
plt.subplot(2, 3, 4)
sns.histplot(df['Marketing Spend'], kde=True, color='orange', bins=15)
plt.title("Marketing Spend Distribution")

# 5. Heatmap: Correlation Matrix
plt.subplot(2, 3, 5)
corr = df[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix")

plt.tight_layout()
plt.show()
```

```python
from dash import Dash, html, dcc
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
# Create visualizations
# 1. Scatter Plot
scatter_fig = px.scatter(df, x='R&D Spend', y='Profit', color='State',
                        title='R&D Spend vs Profit',
                        labels={'R&D Spend': 'R&D Spend ($)', 'Profit': 'Profit ($)'})

# 2. Average Profit by State
avg_profit = df.groupby('State')['Profit'].mean().reset_index()
bar_fig = px.bar(avg_profit, x='State', y='Profit',
                title='Average Profit by State',
                labels={'Profit': 'Average Profit ($)'})

# 3. Profit Distribution
box_fig = px.box(df, y='Profit', title='Profit Distribution',
                labels={'Profit': 'Profit ($)'})

# 4. Marketing Spend Distribution
hist_fig = px.histogram(df, x='Marketing Spend', nbins=20,
                       title='Marketing Spend Distribution',
                       labels={'Marketing Spend': 'Marketing Spend ($)'})

# 5. Correlation Heatmap
corr_matrix = df[['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']].corr()
heatmap_fig = go.Figure(data=go.Heatmap(
                    z=corr_matrix,
                    x=corr_matrix.columns,
                    y=corr_matrix.columns,
                    colorscale='RdBu',
                    zmin=-1,
                    zmax=1))
heatmap_fig.update_layout(title='Feature Correlation Matrix')

# Create Dash app
app = Dash(__name__, external_stylesheets=['https://codepen.io/chriddyp/pen/bWLwgP.css'])

app.layout = html.Div([
    html.H1('Startups Performance Dashboard', style={'textAlign': 'center', 'color': '#2c3e50'}),

    html.Div([
        html.Div([
            dcc.Graph(figure=scatter_fig)
        ], className='six columns'),

        html.Div([
            dcc.Graph(figure=bar_fig)
        ], className='six columns')
    ], className='row'),

    html.Div([
        html.Div([
            dcc.Graph(figure=box_fig)
        ], className='six columns'),

        html.Div([
            dcc.Graph(figure=hist_fig)
        ], className='six columns')
    ], className='row'),

    html.Div([
        html.Div([
            dcc.Graph(figure=heatmap_fig)
        ], className='twelve columns')
    ], className='row'),

    html.Div([
        html.P('Key Insights:', style={'fontWeight': 'bold', 'color': '#2c3e50'}),
        html.Ul([
            html.Li('R&D spending shows strong positive correlation with profits (0.92)'),
            html.Li('New York startups have highest average profit'),
            html.Li('Marketing spend distribution is bimodal'),
            html.Li('Administration costs show no correlation with profits')
        ])
    ], style={'margin': '20px', 'padding': '15px', 'border': '1px solid #eee'})
])

if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```
